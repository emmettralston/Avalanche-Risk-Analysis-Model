# 01 — SNOTEL Exploration

Explore SNOTEL station metadata and raw time series for PNW stations.
Also load and inspect the OAP label archive to understand the label schema.

**Inputs (must exist before running):**
- `data/raw/snotel/snotel_raw.parquet` — produced by `src/data/snotel.py`
- `data/labels/CleanedForecastsNWAC_CAIC_UAC_CAC.V1.2013-2021.zip`
- `configs/zones.geojson`

**Goals:**
1. Map PNW SNOTEL stations against NWAC zone boundaries.
2. Inspect time series completeness (missing data heatmap by station × year).
3. Load OAP labels, filter to NWAC, check class distribution.
4. Quick sanity check: are SNOTEL features even weakly predictive of danger rating?
