# 02 — Terrain Features

Download 3DEP DEMs for each NWAC zone and compute terrain features.

**Inputs (must exist before running):**
- `configs/zones.geojson`

**Outputs:**
- `data/raw/terrain/*.tif` — per-zone DEM and derived rasters
- `data/processed/terrain_features.parquet`

**Goals:**
1. Download 10m DEMs for each zone via py3dep.
2. Compute and visualise slope, aspect, curvature, TRI rasters.
3. Extract zone-level summary statistics.
4. Inspect distribution of steep terrain (slope > 30°) and lee aspects across zones.
