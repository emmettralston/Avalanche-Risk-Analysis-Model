# 05 — Baseline Model

Train and evaluate XGBoost and LightGBM classifiers on the fused feature matrix.

**Inputs (must exist before running):**
- `data/processed/feature_matrix.parquet`
- `data/processed/labels_aligned.parquet`

**Goals:**
1. Weather-only baseline: SNOTEL features alone — establishes floor.
2. Weather + terrain: quantify lift from static terrain features.
3. Full model: weather + terrain + MODIS — quantify satellite lift.
4. Season-aware cross-validation (5-fold); report per-class F1 and kappa.
5. Final evaluation on held-out test seasons (2019-20, 2020-21).
6. Feature importance analysis.
7. Error analysis: where does the model fail? (high-danger misses are costly)
