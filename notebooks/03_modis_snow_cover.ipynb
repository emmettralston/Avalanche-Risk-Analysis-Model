# 03 — MODIS Snow Cover

Download and explore MODIS MOD10A1 / MYD10A1 daily snow cover for NWAC zones.

**Inputs (must exist before running):**
- `configs/zones.geojson`
- NASA Earthdata credentials in `~/.netrc` or environment variables

**Outputs:**
- `data/raw/modis/` — downloaded HDF granules
- `data/processed/modis_snow_cover.parquet`

**Goals:**
1. Authenticate with NASA Earthdata via earthaccess.
2. Search and download granules for one test zone/season.
3. Visualise raw snow cover maps and cloud mask.
4. Quantify cloud cover frequency by zone and month.
5. Evaluate forward-fill vs interpolation strategies for cloudy days.
6. Aggregate to zone-level daily snow cover fraction.
