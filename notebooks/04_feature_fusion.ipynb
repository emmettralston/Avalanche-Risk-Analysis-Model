# 04 — Feature Fusion

Align SNOTEL, terrain, and MODIS modalities to the (date, zone) label index.

**Inputs (must exist before running):**
- `data/raw/snotel/snotel_raw.parquet`
- `data/processed/terrain_features.parquet`
- `data/processed/modis_snow_cover.parquet`
- `data/labels/` — OAP label file
- `configs/zones.geojson`

**Outputs:**
- `data/processed/feature_matrix.parquet` — X (features)
- `data/processed/labels_aligned.parquet` — y (danger ratings, aligned index)

**Goals:**
1. Spatially join stations to zones; inspect zone coverage (stations per zone).
2. Build zone-day weather aggregates and check for missing date ranges.
3. Merge terrain features and confirm no zone mismatches.
4. Merge satellite features; compare coverage with and without MODIS.
5. Inspect feature correlations with danger rating.
6. Verify train/test split by season is leak-free.
